# Blip-2 VS Pure T

```python
import matplotlib.pyplot as plt
import numpy as np

# Metrics
metrics = ["BLEU-1", "BLEU-2", "BLEU-3", "BLEU-4", "CIDEr", "ROUGE-L", "METEOR"]

# Scores for PureT and BLIP-2
puret_scores = [30.00, 13.44, 6.57, 3.67, 15.49, 21.06, 8.60]
blip2_scores = [36.26, 17.60, 8.41, 4.33, 15.19, 22.59, 10.56]

x = np.arange(len(metrics))
width = 0.35

fig, ax = plt.subplots(figsize=(10, 5))
ax.bar(x - width/2, puret_scores, width, label='PureT', color='#4B8BBE')
ax.bar(x + width/2, blip2_scores, width, label='BLIP-2', color='#FFBD33')

ax.set_ylabel('Score')
ax.set_title('Model Comparison: Evaluation Metrics')
ax.set_xticks(x)
ax.set_xticklabels(metrics)
ax.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd

data = {
    "Metric": ["Total Parameters(M)", "GFLOPs", "Inference Time (ms)"],
    "PureT": [229.41, 105.48, 416],
    "BLIP-2": [3700, 761.17, 844]
}

df = pd.DataFrame(data)

# Plot grouped bar chart
x = np.arange(len(df["Metric"]))
width = 0.35

fig, ax = plt.subplots(figsize=(10, 4))
ax.bar(x - width/2, df["PureT"], width, label="PureT", color="#4B8BBE")
ax.bar(x + width/2, df["BLIP-2"], width, label="BLIP-2", color="#FFBD33")

ax.set_ylabel("Value")
ax.set_title("Model Efficiency Comparison")
ax.set_xticks(x)
ax.set_xticklabels(df["Metric"], rotation=15)
ax.legend()
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

# Trade-Offs and Improvements

```python
# --- Metric Comparison ---
metrics = ["BLEU-1", "BLEU-2", "BLEU-3", "BLEU-4", "CIDEr", "ROUGE-L", "METEOR"]
original = [36.26, 17.60, 8.41, 4.33, 15.19, 22.59, 10.56]
finetuned = [54.65, 37.31, 26.76, 19.59, 43.71, 39.13, 26.84]

# Compute % improvement
improvement = [((f - o) / o) * 100 for o, f in zip(original, finetuned)]

# Create DataFrame
df = pd.DataFrame({
    "Metric": metrics,
    "Original": original,
    "Fine-tuned": finetuned,
    "% Improvement": [f"{imp:.2f}%" for imp in improvement]
})

print("Performance Comparison Table:\n")
print(df.to_string(index=False))

df_metrics = pd.DataFrame({
    "Metric": metrics * 2,
    "Score": original + finetuned,
    "Model": ["Original BLIP-2"] * len(metrics) + ["Fine-tuned BLIP-2"] * len(metrics)
})

plt.figure(figsize=(10, 6))
sns.barplot(data=df_metrics, x="Metric", y="Score", hue="Model")
plt.title("Captioning Metrics: Original vs Fine-tuned BLIP-2")
plt.ylim(0, max(finetuned) + 10)
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```
