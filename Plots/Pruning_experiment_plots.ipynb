```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
data = {
    "Model": ["Q6", "Q8", "LLM28", "LLM28+Q8"],
    "BLEU-1": [0.291, 0.50, 0.47, 0.52],
    "BLEU-2": [0.17, 0.312, 0.31, 0.304],
    "BLEU-3": [0.10, 0.20, 0.22, 0.181],
    "BLEU-4": [0.06, 0.129, 0.16, 0.110],
    "METEOR": [0.183, 0.217, 0.215, 0.19],
    "ROUGE": [0.303, 0.36, 0.351, 0.321],
    "CIDEr": [0.304, 0.437, 0.350, 0.371],
    "GFLOPs": [750.58, 750.58, 792.91, 771.75],
    "Inference Time": [508, 456, 2150, 668],
    "Params": [3692, 3709, 3429, 3394]
}

df = pd.DataFrame(data)
```

```python
# Accuracy metrics plot
metrics = ["BLEU-1", "BLEU-2", "BLEU-3", "BLEU-4", "METEOR", "ROUGE", "CIDEr"]
df_melted = df.melt(id_vars="Model", value_vars=metrics, var_name="Metric", value_name="Score")

plt.figure(figsize=(12, 6))
sns.barplot(data=df_melted, x="Metric", y="Score", hue="Model")
plt.title("Caption Quality Metrics Across Models")
plt.ylabel("Score")
plt.xlabel("Evaluation Metric")
plt.legend(title="Model Variant", loc="upper center")
plt.tight_layout()
plt.show()
```

```python
sns.set(style="whitegrid")

# Plot GFlops

plt.figure(figsize=(8, 4))
sns.lineplot(data=df, x="Model", y="GFLOPs", marker="o", color="dodgerblue")
plt.title("GFLOPs per Model")
plt.ylabel("GFLOPs")
plt.xlabel("Model Variant")
plt.tight_layout()
plt.show()
```

```python
# Plot model parameters

plt.figure(figsize=(8, 4))
sns.lineplot(data=df, x="Model", y="Params", marker="s", color="seagreen")
plt.title("Model Parameters")
plt.ylabel("Params (1000M)")
plt.xlabel("Model Variant")
plt.tight_layout()
plt.show()
```

```python
# Plot inference time

plt.figure(figsize=(8, 4))
sns.lineplot(data=df, x="Model", y="Inference Time", marker="^", color="crimson")
plt.title("Inference Time per Image")
plt.ylabel("Time (ms)")
plt.xlabel("Model Variant")
plt.tight_layout()
plt.show()
```
